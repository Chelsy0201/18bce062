```python
'''PRONUNCIATION ACCURACY CALCULATOR'''
'''done by- 18bce062- Chelsy Doshi
            18bce002- Aastha Kumar
            18bce005- Aditya Upadhyay'''
#tkinter is used to make GUI
from tkinter import *
#speech_recognition is used to convert speech to text
import speech_recognition as sr
#gTTs is used to convert text to speech
from gtts import gTTS
#os is used so that we can access inbuilt system-microphone
import os
#pronouncing is used to convert normal words to phonetics 
import pronouncing
#matplotlib.pyplot is used to plot graphical results
import matplotlib.pyplot as plt
#The below two import statements are used to get graph plotted on tkinter window
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure

#list where main input will be stored(string which user wants to enter)
l=[]
#variables used to show how correct their pronunciations were
correct_percent=0
incorrect_percent=0
#initialising the tkinter window
root=Tk()
#customizing tkinter window accoring to our reqirements
root.title("Pronunciation accuracy Calculator")
root.geometry("640x640+450+85")
root.configure(background="white")
heading = Label(root, text="Welcome!", font=("arial", 30, "bold"), fg="orange",bg="white").pack()

label1= Label(root, text="Enter paragraph here : ", font=("arial", 15, "bold"), fg="steelblue",bg="white").place(x=15, y=70)
#This is the actual variable where user entered string is fetched from tkinter window
para= StringVar()
#
entry_box = Entry(root, textvariable= para, width=60, bg="lightgrey").place(x=250,y=80)
#function to take string as input on click
def in_func():
    
    global l
    l.append(str(para.get()))
    Label(root, textvariable=para,font="arial",bg="white",fg="steelblue").place(x=15,y=175)
    Label(root, text="Speak....\nPress start button",font="arial",bg="white",fg="steelblue").place(x=260,y=200)
    
    print(l)
    #return l
#This is the main function which will take speech as input and caculate the accuracy of pronunciation
#also this will return the correct speech of the given input
def main_func():
    global correct_percent
    global incorrect_percent
    l1=[]
    #this list contains input string
    l1.append(str(para.get()))
    
    r = sr.Recognizer()
    #This variable will decide which microphone of system will take input
    mic = sr.Microphone(device_index=0)
    flag=1
    #try-except block will see that system receives proper audio input 
    while(flag):
        print("Speak...")
        
        with mic as source:
            try:
                flag=0
                #this variable actually gets the audio input
                audio = r.listen(source)

            except:
                print("Sorry! was unable to listen...please speak again")
                
                flag=1
                continue
    correct=0 

    
    #this list contains phonetics of words of input string
    l2=[]
    l=l1[0].split()
    for x in l:
        l2.append(pronouncing.phones_for_word(x))
    print(l2)
    #this list contains the text interpreted using speech
    l3=[]
    l3.append(r.recognize_google(audio, language='en'))
    print(l3)
    m=l3[0].split()
    #this list contains the phonetics of speech to text string
    l4=[]
    for y in m:
        l4.append(pronouncing.phones_for_word(y))
    print(l4)
    len_l2=len(l2)
    len_l4=len(l4)
    total=len(l)
    #now we will compare both the phonetics list and we will keep track of no.of correct pronunciated words
    if len_l4<=len_l2:
        for i in range(len_l4):
            if(l2[i]==l4[i]):
                correct=correct+1
            else:
                continue
    print(correct," out of ",total)         
    
    myobj = gTTS(text=l1[0], lang='en', slow=False)
    #this will create text to speech file
    myobj.save("text_to_speech.mp3") 
    #this will create executable audio file
    os.system("start text_to_speech.mp3")
    
    #correct words/ total words
    correct_percent=(correct*100)/total
    incorrect_percent=100-correct_percent
    # Pie chart
    '''labels = ['correct','incorrect']
    sizes = [correct_percent, incorrect_percent]
    explode = (0, 0.1)  
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  
    plt.show()'''
    
#this function will return pie chart of the result on click
def chart_func():
    global correct_percent
    global incorrect_percent
    global pie
    figure = Figure(figsize=(4,3)) 
    subplot = figure.add_subplot() 
    labels = 'correct','incorect' 
    pieSizes = [correct_percent,incorrect_percent]
    
    explode = (0, 0.1)  
    subplot.pie(pieSizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90) 
    subplot.axis('equal')  
    pie = FigureCanvasTkAgg(figure, root)
    pie.get_tk_widget().place(x=165,y=360)
    
#button to take input strig and display it
b1=Button(text="Enter",height=1, width=15, command=in_func).place(x=275,y=120)
#button to run the main function
b2=Button(text="Start",height=1, width=15, command=main_func).place(x=275,y=300)
#button to display results in the form of pie chart
b2=Button(text="Show result",height=1, width=15, command=chart_func).place(x=275,y=320)



root.mainloop()
```
